IMPORT FILE DATA

```python
import numpy as np
import pandas as pd
import os
from apyori import apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
path = 'C:/Users/AnhVu/OneDrive/Máy tính/Đồ Án AI/data/'
```

```python
farmmes = []
for file in os.listdir(path):
    filepath = path + file
    df1 = pd.read_csv(filepath)
    farmmes.append(df1)
    result = pd.concat(farmmes)
df = result
df.to_csv('annualSales2019.csv',index=False)
df
```

Làm sạch dữ liệu.

```python
df['Month'] = df['Order Date'].str[0:2]
df = df.dropna(how='all')
df = df[df['Month'] != 'Or']
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], downcast = 'integer')
df['Order ID'] = pd.to_numeric(df['Order ID'], downcast = 'integer')
df['Price Each'] = pd.to_numeric(df['Price Each'], downcast = 'float')
```

Đếm số lượng sản phẩm.

```python
print('Unique Items: ', df['Product'].nunique())
print('\n',df['Product'].unique())
```

Tạo list mới chứa các sản phẩm của mỗi giao dịch.

```python
transaction_list = []
for i in df['Order ID'].unique():
    list_1 = list(set(df[df['Order ID']==i]['Product']))
    if len(list_1)>0:
        transaction_list.append(list_1)
print(len(transaction_list))
```

```python
transaction_list
```

Mã hóa dữ liệu chuyển về dataframe

```python
te = TransactionEncoder()
te_ary = te.fit(transaction_list).transform(transaction_list)
df2 = pd.DataFrame(te_ary, columns = te.columns_)
df2.sample(10)
```

```python
df3 = (df2.replace(bool == False, '', regex=True))
df3.to_csv('annualSales2019_1.csv',index=False)
```

Tạo bảng luật

```python
frequent_itemsets = apriori(df2, min_support=0.005,use_colnames=True)
rules = association_rules(frequent_itemsets, metric = 'lift',min_threshold=1)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
```

```python
rules[rules['confidence']>0.005].sort_values('confidence', ascending = False)
```
